```python
class Node():
  def __init__(self,point):
    self.point = point
    self.l = None
    self.r = None
```

```python
def insert(root, point, len = 0):
    if root is None:
      return Node(point)
    axis = len % 2
    if point[axis] < root.point[axis]:
      root.l = insert(root.l, point, len+1)
    else:
      root.r = insert(root.r, point, len+1)

    return root
```

```python
def nearestPoint(root, target, len = 0, nearest = None):
  if root is None:
    return nearest

  axis = len % 2
  next_nearest = None
  next_branch = None

  if nearest is None or distance(target,nearest) > distance(target,root.point):
    next_nearest = root.point
  else:
    next_nearest = nearest

  if target[axis] < root.point[axis]:
    next_branch = root.l
  else:
    next_branch = root.r

  return nearestPoint(next_branch,target,len+1,next_nearest)
```

```python
def distance(p1,p2):
  return ((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)**0.5
```

```python
def main():
    root = None

    for i in range(10):
        x = int(input(f"Enter x-coordinate {i + 1}: "))
        y = int(input(f"Enter y-coordinate {i + 1}: "))
        point = (x, y)
        root = insert(root, point)

    new_x =int(input("Enter new x-coordinate : "))
    new_y = int(input("Enter new y-coordinate : "))
    new_point = (new_x, new_y)

    nearest_point = nearestPoint(root, new_point)
    print(f"The nearest point to the new point is: {nearest_point}")


if __name__ == "__main__":
    main()
```
